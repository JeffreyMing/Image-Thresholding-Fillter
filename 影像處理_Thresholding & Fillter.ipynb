```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family']='DFKai-SB' #顯示中文(for Win10)，可自行更換可用字體
plt.rcParams['axes.unicode_minus']=False #正常顯示負號
%matplotlib inline
```

```python
img = cv2.imread('food.jpg')
```

```python
#灰階化
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
```

```python
cv2.imshow('img_gray', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Thresholding 二值化

```python
#二值化
ret, thresh1 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img_gray, 127, 255, cv2.THRESH_TOZERO_INV)

titles = ['Original', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img_gray, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
plt.show()
```

```python
cv2.imshow('thresh1', thresh1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 濾波器Filter

```python
#原圖
img = cv2.imread('noise.jpg')

#均值濾波 Mean Filter
mean = cv2.blur(img, (3,3))
# box filtetr
box = cv2.boxFilter(img, -1,(3,3), normalize=True)
#Gaussian Filter
gauss = cv2.GaussianBlur(img, (5,5),1)
#中值濾波 Median Filter
median = cv2.medianBlur(img, 5)
```

```python
titles = ['Original', 'mean', 'box', 'gauss', 'median']
images = [img, mean, box, gauss, median]
```

```python
#合併顯示
img_hstack = np.hstack((images))
cv2.namedWindow('img_hstack', cv2.WINDOW_FREERATIO)
cv2.imshow('img_hstack', img_hstack)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(12,8))
for i in range(5):
    plt.subplot(2,3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
plt.show()
```
